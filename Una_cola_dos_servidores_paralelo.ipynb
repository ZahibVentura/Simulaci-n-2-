```python
import numpy as np
import random as rd

# Subrutina para generar tiempos de llegada utilizando un proceso no homogéneo
def expo(T1=20, lamda=0.4):
    t = 0
    Lista = []
    while True:
        u1 = rd.uniform(0, 1)
        t = t - np.log(u1) / lamda  # Genera el tiempo de llegada
        if t > T1:
            break
        u2 = rd.uniform(0, 1)
        if u2 <= lamda_t(t) / lamda:  # Verifica si el evento es válido
            Lista.append(t)  # Guarda el tiempo de llegada
    return Lista

# Función que define la tasa de llegada λ(t) en función del tiempo
def lamda_t(t=3):
    if 0 <= t <= 3:
        return 5 + 5 * t
    elif 3 < t <= 5:
        return 20
    elif 5 < t <= 9:
        return 20 - 2 * (t - 5)
    else:
        return 0

# Parámetros iniciales
t = 0  # Tiempo actual
C1 = 0  # Número de clientes servidos por servidor 1
C2 = 0  # Número de clientes servidos por servidor 2
NLL = 0  # Número total de llegadas
SS = [0]  # Estado del sistema (número de clientes en la cola y atendidos por S1 y S2)

# Generación de la primera llegada
T0 = expo()[0]  # Primer tiempo de llegada
t_LL = T0  # Tiempo de la próxima llegada
t1 = np.inf  # Tiempo de la próxima salida del servidor 1
t2 = np.inf  # Tiempo de la próxima salida del servidor 2

# Comienza la simulación
while True:
    # Caso 1: Llega un cliente
    if (t_LL <= t1) and (t_LL <= t2):
        t = t_LL
        NLL += 1  # Actualiza el número de llegadas
        print(f"Llegada del cliente {NLL} en el tiempo {t:.2f}")

        # Generamos la próxima llegada
        Tt = expo()  # Nuevas llegadas
        t_LL = Tt[0] if len(Tt) > 0 else np.inf

        if SS == [0]:
            # Si el sistema está vacío, el cliente entra a S1
            SS = [1, NLL, 0]  # El primer cliente es atendido por S1
            Y1 = np.random.exponential(20)  # Tiempo de servicio para S1
            t1 = t + Y1
        elif SS == [1, NLL, 0]:
            # Si hay un cliente en S1 pero no en S2, este cliente va a S2
            SS = [2, NLL, NLL]
            Y2 = np.random.exponential(20)  # Tiempo de servicio para S2
            t2 = t + Y2
        else:
            # Si hay más de un cliente, se añade a la cola
            SS = [SS[0] + 1] + SS[1:] + [NLL]

    # Caso 2: Sale un cliente de S1
    elif t1 <= t_LL and t1 <= t2:
        t = t1
        C1 += 1  # Actualizamos el número de clientes servidos por S1
        print(f"Salida del cliente atendido por S1 en tiempo {t:.2f}")

        if SS[0] == 1:
            # Si no hay más clientes en el sistema, S1 queda vacío
            SS = [0]
            t1 = np.inf
        elif SS[0] == 2:
            # Si hay un cliente en S2, solo queda ese en el sistema
            SS = [1, 0, SS[2]]
            t1 = np.inf
        else:
            # Si hay más clientes, el siguiente pasa a S1
            SS = [SS[0] - 1] + [SS[2]] + SS[3:]
            Y1 = np.random.exponential(20)  # Tiempo de servicio del próximo cliente
            t1 = t + Y1

    # Caso 3: Sale un cliente de S2
    elif t2 <= t_LL and t2 <= t1:
        t = t2
        C2 += 1  # Actualizamos el número de clientes servidos por S2
        print(f"Salida del cliente atendido por S2 en tiempo {t:.2f}")

        if SS[0] == 1:
            # Si no hay más clientes en el sistema, S2 queda vacío
            SS = [0]
            t2 = np.inf
        elif SS[0] == 2:
            # Si hay un cliente en S1, solo queda ese en el sistema
            SS = [1, SS[1], 0]
            t2 = np.inf
        else:
            # Si hay más clientes, el siguiente pasa a S2
            SS = [SS[0] - 1, SS[1]] + SS[3:]
            Y2 = np.random.exponential(20)  # Tiempo de servicio del próximo cliente
            t2 = t + Y2

    # Si ambos tiempos de salida son mayores al tiempo límite, terminamos
    if t_LL == np.inf and t1 == np.inf and t2 == np.inf:
        break

# Imprime los resultados finales
print(f"Número total de clientes servidos por S1: {C1}")
print(f"Número total de clientes servidos por S2: {C2}")
```
