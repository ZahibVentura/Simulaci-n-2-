```python
import numpy as np
import random as rd

# Subrutina lamda_t
def lamda_t(t=3):
    if 0 <= t <= 3:
        z = 5 + 5 * t
    elif 3 < t <= 5:
        z = 20
    elif 5 < t <= 9:
        z = 20 - 2 * (t - 5)
    else:
        z = 0
    return z

# Subrutina expo
def expo(T1=20, lamda=0.4):
    t = 0
    I = 0
    Lista = []
    while True:
        u1 = rd.uniform(0, 1)
        t = t - np.log(u1) / lamda
        if t > T1:
            break
        u2 = rd.uniform(0, 1)
        if u2 <= lamda_t(t) / lamda:
            I += 1
            Lista.append(t)
    return Lista

# Inicialización de variables
t = 0
n1 = 0
n2 = 0
NLL = 0
Ns = 0

# Tiempo total de simulación (puedes ajustarlo)
Tto = 20

# Generar tiempos de llegada con la subrutina expo
To = expo(Tto)
if To:  # Verificar si la lista no está vacía
    t_LL = To[0]  # Primer tiempo de llegada
else:
    t_LL = np.inf  # Si no hay llegadas, establecer t_LL en infinito
#t_LL = To[0]  # Primera llegada al servidor 1

t1 = np.inf  # Tiempo de salida de servidor 1
t2 = np.inf  # Tiempo de salida de servidor 2

LL1 = []  # Lista de llegadas al servidor 1
LL2 = []  # Lista de llegadas al servidor 2
S = []    # Lista de salidas del sistema

# Inicio de la simulación
while True:
    if t_LL <= t1 and t_LL <= t2:
        # Caso 1: Llegada de un cliente al servidor 1
        t = t_LL
        NLL += 1
        n1 += 1
        LL1.append(t)  # Anotar la llegada al servidor 1
        Tt = expo(Tto)
        if Tt:  # Verificar si la lista no está vacía
            t_LL = Tt[0]  # Actualizamos t_LL al siguiente tiempo de llegada
        else:
            t_LL = np.inf  # Si no hay más llegadas, establecer t_LL en infinito
        #t_LL = Tt[0]  # Nueva llegada al servidor 1
        if n1 == 1:
            Y1 = rd.uniform(0, 1)  # Tiempo de servicio del servidor 1
            t1 = t + Y1

    elif t1 < t_LL and t1 <= t2:
        # Caso 2: Salida de un cliente del servidor 1 hacia el servidor 2
        t = t1
        n1 -= 1
        n2 += 1
        LL2.append(t)  # Anotar la llegada al servidor 2
        if n1 == 0:
            t1 = np.inf  # No hay más clientes en el servidor 1
        else:
            Y1 = rd.uniform(0, 1)
            t1 = t + Y1  # Próxima salida del servidor 1
        if n2 == 1:
            Y2 = rd.uniform(0, 1)  # Tiempo de servicio del servidor 2
            t2 = t + Y2

    elif t2 < t_LL and t2 < t1:
        # Caso 3: Salida de un cliente del servidor 2 (salida del sistema)
        t = t2
        Ns += 1
        n2 -= 1
        S.append(t)  # Anotar la salida del sistema
        if n2 == 0:
            t2 = np.inf  # No hay más clientes en el servidor 2
        else:
            Y2 = rd.uniform(0, 1)
            t2 = t + Y2  # Próxima salida del servidor 2

    if t_LL > Tto and t1 == np.inf and t2 == np.inf:
        # Todos los eventos han terminado, terminar simulación
        break

# Imprimir resultados
print("Instantes de llegada al servidor 1 (LL1):", LL1)
print("Instantes de llegada al servidor 2 (LL2):", LL2)
print("Instantes de salida del sistema (S):", S)

# Cálculos adicionales solicitados
# Tiempo medio que un cliente pasa en el sistema
if NLL > 0:
    tiempo_medio_sistema = np.mean([S[i] - LL1[i] for i in range(len(S))])
else:
    tiempo_medio_sistema = 0

# Número medio de clientes en el sistema
tiempo_total = max(S) if len(S) > 0 else 0
numero_medio_sistema = NLL / tiempo_total if tiempo_total > 0 else 0

# Tiempo después de cierre (diferencia entre el cierre y la última salida)
tiempo_despues_cierre = max(0, max(S) - Tto) if len(S) > 0 else 0

# Última salida (instante de tiempo del último cliente)
ultima_salida = max(S) if len(S) > 0 else 0

print("Tiempo medio que un cliente pasa en el sistema:", tiempo_medio_sistema)
print("Número medio de clientes en el sistema:", numero_medio_sistema)
print("Tiempo después de cierre:", tiempo_despues_cierre)
print("Última salida del sistema:", ultima_salida)
```
