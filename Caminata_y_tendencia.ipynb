```python
import random as rd
import matplotlib.pyplot as plt
```

```python
def caminata_aleatoria(p,N,contador):
  Lista = [contador]
  for i in range(N):
    tiro = rd.uniform(0,1)
    if tiro <= p:
      contador += 1
    else:
      contador -= 1
    Lista.append(contador)
  return Lista
```

```python
L_Promedio = []
for j in range(100):
  Lista = caminata_aleatoria(0.5,400,0)
  #L_Promedio.append(sum(Lista)/len(Lista))
  plt.plot(Lista)
plt.show()
```
