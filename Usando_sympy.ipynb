```python
import sympy as sp
import numpy as np

sp.init_printing()
```

```python
S,t,c = sp.symbols('S,t,c')
```

```python
(t+4)**2
```

```python
((t+4)**2).expand()
```

```python
((t+4)**2).subs(t,S)
```

```python
sp.factor(t**2 + 6*t + 5)
```

```python
sp.simplify(t**2 + 2*t + 5+2*t**2)
```

```python
sp.plotting.plot(t*t)
```

```python
sp.plotting.plot(0.5*sp.exp(0.1*t))
```

```python
S = sp.Function('S')
ode = sp.Eq(sp.Derivative(S(t),t)-0.1*S(t),0)
ode
```

```python
sol = sp.dsolve(ode,S(t),ics = {S(0):1000})
sol
```

```python
sp.plotting.plot(sol.rhs)
```

```python
sp.plotting.plot(sp.sin(t))
```

```python
Y = sp.Heaviside(t-sp.pi)
sp.plotting.plot(Y)
```

```python
Esca = sp.cos(t)+ (-sp.cos(t)+sp.sin(t))*sp.Heaviside(t-sp.pi)
sp.plotting.plot(Esca)
```

```python
Es =  (1/2)*sp.Heaviside(t-2) - (1/2)*sp.Heaviside(t-4)
sp.plotting.plot(Es, (t,0,5))
```

```python
D = sp.DiracDelta(t-3)
sp.plotting.plot(D)
```

```python
NN = t*(sp.Heaviside(t)-sp.Heaviside(t-1))
sp.plotting.plot(NN, xlim=(-1,2))
```
