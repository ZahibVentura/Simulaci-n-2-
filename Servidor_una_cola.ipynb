```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
def lamda_t(t =3):
  if 0 <= t <= 3:
    z = 5 + 5*t
  elif 3 < t <= 5:
    z = 20
  elif 5 < t <= 9:
    z = 20 - 2*(t-5)
  else:
    z = 0
  return z
```

```python
def expo(T1 = 20, lamda = 0.4):
  t = 0
  I = 0
  Lista = []
  while True:
    u1 = rd.uniform(0,1)
    t = t - np.log(u1)/lamda
    if t > T1:
      break
    u2 = rd.uniform(0,1)
    if u2 <= lamda_t(t)/lamda:
      I += 1
      Lista.append(t)
  return Lista

print(expo())
```
