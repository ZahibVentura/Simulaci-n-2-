```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt

# Función lambda_t que devuelve la tasa de llegada en función del tiempo t
def lamda_t(t=3):
    if 0 <= t <= 3:
        z = 5 + 5*t
    elif 3 < t <= 5:
        z = 20
    elif 5 < t <= 9:
        z = 20 - 2*(t-5)
    else:
        z = 0
    return z

# Función expo para generar tiempos de llegada según la tasa de llegada variable lambda_t
def expo(T1=20, lamda=0.4):
    t = 0
    Lista = []
    while True:
        u1 = rd.uniform(0, 1)
        t = t - np.log(u1) / lamda
        if t > T1:
            break
        u2 = rd.uniform(0, 1)
        if u2 <= lamda_t(t) / lamda:
            Lista.append(t)
    return Lista

# VARIABLES DE INICIO
Tto = 9  # Tiempo de cierre (Tp) = 9 horas
t = 0    # Tiempo actual
n = 0    # Número de clientes en el sistema
N_LL = 0 # Número de llegadas hasta el instante t
N_s = 0  # Número de salidas hasta el instante t


# Generamos la primera llegada con la función expo
To = expo(Tto)
if To:  # Verificar si la lista no está vacía
    t_LL = To[0]  # Primer tiempo de llegada
else:
    t_LL = np.inf  # Si no hay llegadas, establecer t_LL en infinito
LL = []  # Lista para almacenar tiempos de llegada
S = []   # Lista para almacenar tiempos de salida

# Inicializamos el tiempo de salida en infinito (no hay clientes aún)
ts = np.inf

# Bucle para simular las llegadas y salidas de clientes
while True:
    # Caso 1: Llegada de un cliente
    if t_LL <= ts and t_LL <= Tto:
        t = t_LL  # Actualizamos el tiempo al de llegada
        N_LL += 1  # Incrementamos el contador de llegadas
        n += 1  # Aumenta el número de clientes en el sistema
        LL.append(t)  # Registramos la llegada en la lista
        Tt = expo(Tto)  # Generamos el siguiente tiempo de llegada
        if Tt:  # Verificar si la lista no está vacía
            t_LL = Tt[0]  # Actualizamos t_LL al siguiente tiempo de llegada
        else:
            t_LL = np.inf  # Si no hay más llegadas, establecer t_LL en infinito
        if n == 1:
            Y = rd.uniform(0, 1)  # Tiempo de atención del cliente
            ts = t + Y  # Actualizamos el tiempo de salida del cliente
    # Caso 2: Salida de un cliente
    elif ts < t_LL and ts <= Tto:
        t = ts  # Actualizamos el tiempo al de salida
        N_s += 1  # Incrementamos el contador de salidas
        n -= 1  # Decrementamos el número de clientes en el sistema
        S.append(t)  # Registramos la salida en la lista
        if n == 0:
            ts = np.inf  # No hay más clientes en el sistema
        else:
            Y = rd.uniform(0, 1)  # Tiempo de atención del siguiente cliente
            ts = t + Y  # Actualizamos el tiempo de salida
    # Caso 3: El sistema sigue atendiendo después del cierre
    elif ts > Tto and t_LL > Tto and n > 0:
        t = ts  # Actualizamos el tiempo al de salida
        N_s += 1  # Incrementamos el contador de salidas
        n -= 1  # Decrementamos el número de clientes en el sistema
        S.append(t)  # Registramos la salida en la lista
        if n > 0:
            Y = rd.uniform(0, 1)  # Tiempo de atención del siguiente cliente
            ts = t + Y  # Actualizamos el tiempo de salida
        else:
            ts = np.inf  # No hay más clientes en el sistema
    # Caso 4: El sistema cierra y no hay más clientes en espera
    elif ts > Tto and t_LL > Tto and n == 0:
        T_p = np.max([0, t - Tto])  # Calculamos el tiempo adicional de cierre
        break  # Terminamos la simulación

# Resultados finales
print("Instantes de llegada de los clientes: ", LL)
print("Instantes de salida de los clientes: ", S)




# 1. Tiempo medio de un cliente en el sistema
if N_LL > 0:
    tiempo_medio_cliente = np.mean([S[i] - LL[i] for i in range(N_LL)])
    print("Tiempo medio de un cliente en el sistema: ", tiempo_medio_cliente)
else:
    print("No hubo llegadas de clientes.")

# 2. Número medio de clientes en el sistema
tiempo_total = Tto  # Tiempo de jornada (hasta el cierre)
numero_medio_clientes = N_LL / tiempo_total if tiempo_total > 0 else 0
print("Número medio de clientes en el sistema: ", numero_medio_clientes)

# 3. Tiempo después del cierre
print("Tiempo después de cierre: ", T_p)

# 4. Última salida
if S:
    ultima_salida = S[-1]
    print("Última salida: ", ultima_salida)
else:
    print("No hubo salidas de clientes.")
```
