###Serpientes y escaleras

 ¿Cuál es, en promedio, el número de tiradas necesarias para terminar el juego? Considerar que cada tirada se hace con un dado ordinario.


Responder:

a) Analíticamente.

b) Por simulación.


Serpientes y escaleras.png


```python
"""Por simulación"""

import random as rd    #importamos las librerias a utilizar
import numpy as np

lista = []
lista_promedios = [] #definimos dos listas, una tendra el numero de tiradas por juego y otra
N = 10000            #el esperado de tiradas por cada vez que ejecutemos el experimento


for j in range(10):   #ciclo para encontrar el esperado 10 veces y compararlos
  for i in range(N):  #ciclo para calcular un solo esperado
    contador = 0
    suma = 0
    while suma <20:   #condicion del numero de casillas limites del tablero (al llegar a 20 0 mas ganas)
      dado = rd.randint(1,6)  #simulamos nuestro dado de 6 caras
      suma += dado
      contador +=1
      if suma == 3:  #condicion de escalera 3 al 11
        suma +=8
      elif suma == 15:  #condicion de escalera 15 al 19
        suma +=4
      elif suma == 13:  #condicion de serpiente 13 al 4
        suma -=9
      elif suma == 17:  #condicion de serpiente 17 al 10
        suma -=7
    lista.append(contador)  #agregamos a la lista el numero de tiradas por juego
  promedio = np.mean(lista)  #calculamos los promedios
  lista_promedios.append(promedio)

print('promdeio de 10 veces el experimento: ',lista_promedios)
print('El promedio de tiradas necesarias para terminar el juego es: ',np.mean(lista_promedios))
```
